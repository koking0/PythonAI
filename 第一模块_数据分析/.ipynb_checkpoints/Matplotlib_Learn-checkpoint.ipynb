# Matplotlib

Matplotlib是一个强大的Python绘图和数据可视化的工具包。

plot函数：绘制折线图

    线型linestyle(-, -., --, ...)
    
    点型marker(v, ^, s, *, H, +, x, D, o, ...)
    
    颜色color(b, g, r, y, k, w, ...)
    
plot函数绘制多条曲线

pandas包对plot的支持

```python
import matplotlib.pyplot as plt
```

```python
plt.plot()
plt.show()
```

```python
plt.plot([1,2,3,4], [9,8,7,6])
plt.show()
```

```python
plt.plot([1,2,3,4], [9,8,7,6], "o")
plt.show()
```

```python
plt.plot([1,2,3,4], [9,8,7,6], "o--k")
plt.show()
```

```python
plt.plot([1,2,3,4], [9,8,7,6], "v-.y")
plt.show()
```

```python
plt.plot([1,2,3,4], [9,8,7,6], "*-b")
plt.show()
```

```python
plt.plot([1,2,3,4], [9,8,7,6], "D-r")
plt.show()
```

## 图像标注

| 方法 | 功能 |
| --- | --- |
| plt.title() | 设置图像标题 |
| plt.xlabel() | 设置x轴名称 |
| plt.ylabel() | 设置y轴名称 |
| plt.xlim() | 设置x轴范围 |
| plt.tlim() | 设置y轴范围 |
| plt.xticks() | 设置x轴刻度 |
| plt.yticks() | 设置y轴刻度 |
| plt.legend() | 设置曲线图例 |

```python
import numpy as np

plt.plot([np.random.randint(0, 10) for _ in range(5)], [np.random.randint(0, 10) for _ in range(5)], color="black", label="Line A")
plt.plot([np.random.randint(0, 10) for _ in range(5)], [np.random.randint(0, 10) for _ in range(5)], color="green", label="Line B")
plt.title("Matplotlib Test Plot")
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.xlim(0, 11)
plt.ylim(0, 11)
plt.legend()
plt.show()
```

### Matplotlib实例——绘制数学函数

```python
x = np.linspace(-100, 100, 1000)
y1 = x
y2 = x ** 2
y3 = np.sin(x)
plt.plot(x, y1, color="red", label="y1")
plt.plot(x, y2, color="blue", label="y2")
plt.plot(x, y3, color="green", label="y3")
plt.xlim(0, 100)
plt.ylim(-10, 100)
plt.legend()
plt.show()
```
